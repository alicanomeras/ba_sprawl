Berechnung f√ºr die zeitliche Entwicklung

```python
def graph(df1, df2, df3, indicator, title , titel_indic, color, plot_name):

    font_leg = {'family': 'helvetica', 'size': 10}
    font_tit = {'family': 'helvetica', 'size': 12}
    
    values = [df1[indicator], df2[indicator], df3[indicator] ]

    years = [2015, 2020, 2021]

    fig, ax = plt.subplots(figsize=(7,5))

    ax.plot(years, values, color=color)
    
    plt.xlabel('Jahre', fontdict=font_leg)
    plt.ylabel(titel_indic, fontdict=font_leg)
    plt.title('Entwicklung der ' + title, fontdict=font_tit, pad=15)

    #fig.savefig('/Users/petole/gitlab/bachelor-arbeit/graphen/{}.png'.format(plot_name))

    return plt.show()
```

Graph f√ºr Vergleich

```python
import matplotlib.ticker as ticker
def graph_comp(df11, df12, df13, df21, df22, df23, df31, df32, df33, df41, df42, df43, indicator, title , titel_indic, plot_name):

    font_leg = {'family': 'helvetica', 'size': 10}
    font_tit = {'family': 'helvetica', 'size': 12}

    region1 = [df11[indicator].iloc[0], df12[indicator].iloc[0], df13[indicator].iloc[0] ]
    region2 = [df21[indicator].iloc[0], df22[indicator].iloc[0], df23[indicator].iloc[0] ]
    region3 = [df31[indicator].iloc[0], df32[indicator].iloc[0], df33[indicator].iloc[0] ]
    region4 = [df41[indicator].iloc[0], df42[indicator].iloc[0], df43[indicator].iloc[0] ]

    years = [2015, 2020, 2021]

    fig, ax = plt.subplots(figsize=(8,5))

    proz_1 = ((df13[indicator]-df11[indicator])*100/df11[indicator]).iloc[0]
    proz_2 = ((df23[indicator]-df21[indicator])*100/df21[indicator]).iloc[0]
    proz_3 = ((df33[indicator]-df31[indicator])*100/df31[indicator]).iloc[0]
    proz_4 = ((df43[indicator]-df41[indicator])*100/df41[indicator]).iloc[0]

    proz = [proz_1, proz_2, proz_3, proz_4]

    for i, line in enumerate([region1, region2, region3, region4]):
        x_value = max(years)
        y_value = line[-1]
        ax.text(x_value + 0.02*(max(years)-min(years)), y_value + 0.02*(max(line)-min(line)), f'{proz[i]:.2f}%')


    plt.xlim(min(years), max(years) + 0.1*(max(years)-min(years)))
    ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

    ax.plot(years, region1, label='Kanton Obwalden', color='#7fc97f')
    ax.plot(years, region2, label='Kanton Glarus', color='#beaed4')
    ax.plot(years, region3, label='Region Bernina', color='#fdc086')
    ax.plot(years, region4, label='Region Unterengadin und Val M√ºstair', color='#e78ac3')

    plt.xlabel('Jahre', fontdict=font_leg)
    plt.ylabel(titel_indic, fontdict=font_leg)
    plt.title('Entwicklung der ' + title, fontdict=font_tit, pad=15)
    ax.legend(fontsize=8, loc='lower left', bbox_to_anchor=(0.01, 0.36))
    fig.savefig('/Users/petole/gitlab/bachelor-arbeit/graphen/{}.png'.format(plot_name))

    return plt.show()
```

Fl√§chenverbrauch (Gebiets-, Besiedelbare und besiedelte Fl√§che)

```python
font_leg = {'family': 'helvetica', 'size': 10}
font_tit = {'family': 'helvetica', 'size': 12}

#convert in m^2 -> ha 
set_OB = OB_2015['settlement_area']/10000 #ha
set_GL = GL_2020['settlement_area']/10000
set_VP = reg_VP_2021['settlement_area']/10000
set_UE_VM = reg_UE_VM_2021['settlement_area']/10000

gem_surf_ha_VP = (reg_VP_2021.area)/10000
gem_surf_ha_UE_VM = (reg_UE_VM_2021.area)/10000
gem_surf_ha_OB = (OB_2021.area)/10000
gem_surf_ha_GL = (GL_2021.area)/10000

SSA = [OB_2021['tot_SSA_surf'].iloc[0], GL_2021['tot_SSA_surf'].iloc[0], reg_VP_2021['tot_SSA_surf'].iloc[0], reg_UE_VM_2021['tot_SSA_surf'].iloc[0]] #besiedelbar #anstatt help2 = GL_2021['SSA_Surf_ha'].iloc[0]
settlement = [set_OB.iloc[0], set_GL.iloc[0], set_VP.iloc[0], set_UE_VM.iloc[0]] #siedlung 
flaeche = [gem_surf_ha_OB.iloc[0], gem_surf_ha_GL.iloc[0], gem_surf_ha_VP.iloc[0], gem_surf_ha_UE_VM.iloc[0]] #gebiet

gebiete = ['Obwalden', 'Glarus', 'Valposchiavo', 'Unterengadin und Val M√ºstair']
x_axis = np.arange(len(gebiete))

fig, ax = plt.subplots(figsize=(7,5))

share = [(OB_2021['tot_share']*100).iloc[0], (GL_2021['tot_share']*100).iloc[0], (reg_VP_2021['tot_share']*100).iloc[0], (reg_UE_VM_2021['tot_share']*100).iloc[0]]

for i in range(len(gebiete)):
    ax.text(x_axis[i], flaeche[i]+1000, f'{share[i]:.2f}%', ha='center')

plt.bar(x_axis, np.array(SSA), 0.2, label='Besiedelbare Fl√§che', color='#e78ac3')
plt.bar(x_axis-0.2, np.array(settlement), 0.2, label='Siedlungsfl√§che', color='#fdc086')
plt.bar(x_axis+0.2, np.array(flaeche), 0.2, label='Gebietsfl√§che', color='#7fc97f')
ax.legend(fontsize=8, loc='best')

plt.xlabel('Untersuchungsgebiete', fontdict=font_leg) 
plt.ylabel('Fl√§che [ha]', fontdict=font_leg)
plt.xticks(x_axis, gebiete)
ax.set_xticklabels(gebiete, fontsize=9)
plt.title('Fl√§chenverbrauch in den vier Untersuchungsgebieten im Jahr 2021', fontdict=font_tit, pad=15)

fig.savefig('/Users/petole/gitlab/bachelor-arbeit/graphen/flaeche_vergleich.png')

plt.show()
```

```python
import matplotlib.ticker as ticker
def graph_comp_bauzonen(df11, df12, df21, df22, df31, df32, indicator, title , titel_indic, plot_name):

    font_leg = {'family': 'helvetica', 'size': 10}
    font_tit = {'family': 'helvetica', 'size': 12}

    gem = [df11[indicator].iloc[0], df12[indicator].iloc[0]]
    in_bau = [df21[indicator].iloc[0], df22[indicator].iloc[0]]
    aus_bau = [df31[indicator].iloc[0], df32[indicator].iloc[0]]

    years = [2015, 2020]

    fig, ax = plt.subplots(figsize=(8,5))

    proz_1 = ((df12[indicator]-df11[indicator])*100/df11[indicator]).iloc[0]
    proz_2 = ((df22[indicator]-df21[indicator])*100/df21[indicator]).iloc[0]
    proz_3 = ((df32[indicator]-df31[indicator])*100/df31[indicator]).iloc[0]

    proz = [proz_1, proz_2, proz_3]

    for i, line in enumerate([gem, in_bau, aus_bau]):
        x_value = max(years)
        y_value = line[-1]
        ax.text(x_value + 0.02*(max(years)-min(years)), y_value + 0.02*(max(line)-min(line)), f'{proz[i]:.2f}%')


    ax.plot(years, gem, label='Gemeinde', color='#7fc97f')
    ax.plot(years, in_bau, label='Innerhalb der Bauzone', color='#beaed4')
    ax.plot(years, aus_bau, label='Ausserhalb der Bauzone', color='#fdc086')

    plt.xlim(min(years), max(years) + 0.1*(max(years)-min(years)))
    ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

    plt.xlabel('Jahre', fontdict=font_leg)
    plt.ylabel(titel_indic, fontdict=font_leg)
    plt.title('Entwicklung der ' + title, fontdict=font_tit, pad=15)
    ax.legend(fontsize=8, loc='center left', bbox_to_anchor=(0.01, 0.7))

    #fig.savefig('/Users/petole/gitlab/bachelor-arbeit/graphen/{}.png'.format(plot_name))

    return plt.show()
```

```python
def hist_diff(df11, df13, df21, df23, df31, df33, df41, df43, indicator, title , titel_indic, plot_name):
    
    font_leg = {'family': 'helvetica', 'size': 10}
    font_tit = {'family': 'helvetica', 'size': 12}

    gebiete = ['Obwalden', 'Glarus', 'Valposchiavo', 'Unterengadin/Val M√ºstair']

    fig, ax = plt.subplots(figsize=(8,5))

    diff_1 = (df13[indicator]-df11[indicator]).iloc[0]
    diff_2 = (df23[indicator]-df21[indicator]).iloc[0]
    diff_3 = (df33[indicator]-df31[indicator]).iloc[0]
    diff_4 = (df43[indicator]-df41[indicator]).iloc[0]

    diff = [diff_1, diff_2, diff_3, diff_4]

    start_values = [df11[indicator].iloc[0], df21[indicator].iloc[0], df31[indicator].iloc[0], df41[indicator].iloc[0]]

    #ax.set_ylim(-1,1)


    colors = ['#7fc97f', '#beaed4', '#fdc086', '#e78ac3']

    bar_container = ax.bar(gebiete, diff, color=colors, width=0.8)
    ax.axhline(0, color='grey', linewidth=0.5)

    for rect, start_value in zip(bar_container, start_values):
        height = rect.get_height()
        if height > 0: # if the bar is above x-axis
            position = 0.01 #height - 0.02
        else: # if the bar is below x-axis
            position = height + 0.001
        ax.text(rect.get_x() + rect.get_width() / 2, position, 'Anfangswert\n' +
                '{:.2f}'.format(start_value), # format height to 2 decimal places
                ha='center', va='bottom')

    plt.xlabel('Untersuchungsgebiete', fontdict=font_leg, labelpad=12)
    plt.ylabel(titel_indic, fontdict=font_leg)
    plt.title('Absolute Differenz der ' + title + '\nzwischen den Jahren 2015 und 2021', fontdict=font_tit, pad=15)

    fig.savefig('/Users/petole/gitlab/bachelor-arbeit/graphen/{}.png'.format(plot_name))
    
    return plt.show()
```
